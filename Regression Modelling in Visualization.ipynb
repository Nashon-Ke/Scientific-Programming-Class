# Regression Modelling in Visualization

## Data Visualization

Matplotlib is the most popular python package for twenty graphics. It provides a very quick way to visualize data from python and publication quality figures in many formats.

### Matplotlib in pylab

This provides nice tools for making plots od 1D and 2D data (curves, contour, plots e.t.c).
The best way to get matplotlib working interactively in a standard python shell is to:

```python
import numpy as np
import matplotlib.pyplot as plt
plt.interactive(True)
```

```python
x=np.linspace(-1,1,20)
```

```python
x
```

```python
plt.plot(x,x**2,'o--')
```

```python
plt.plot(x,x**2,'o--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('y=x^2')
plt.grid(True)
plt.show
```

### Labels, Titles, Grids, Overplots and Legends

We can use greek letters as in LaTex. You should use raw strings(precode the quotes with r) and surround the math text with dollar sign($).

```python
plt.plot(x,x**2,'o--')
plt.xlabel(r'$\alpha$')
plt.ylabel(r'$y=x^2$')
plt.title(r'$y=x^2$')
plt.grid(True)
```

We can plot more than one function in the same figure.

```python
plt.interactive(False)
plt.figure(figsize=(8,6), dpi=80) #Converted a figure of size 8x6 inches, and 80 dots per inch
x=np.linspace(-np.pi, np.pi, 256, endpoint=True)
c,s=np.cos(x), np.sin(x)
plt.plot(x,c, color='blue', linewidth=3.0, linestyle='--')
plt.plot(x,s, color='green', linewidth=3.0, marker='d', markersize=5)
plt.grid(True)
```

Labelling the graph

```python
plt.interactive(False)
plt.figure(figsize=(8,6), dpi=80) #Converted a figure of size 8x6 inches, and 80 dots per inch
x=np.linspace(-np.pi, np.pi, 256, endpoint=True)
c,s=np.cos(x), np.sin(x)
plt.plot(x,c, color='blue', linewidth=3.0, linestyle='--')
plt.plot(x,s, color='green', linewidth=3.0, marker='d', markersize=5)
plt.grid(True)
plt.xlabel('$x$')
plt.ylabel(r'$y=x^2$')
plt.xlim(-4.0,4.0)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi, np.pi/2, np.pi], [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi$', r'$+\pi/2$'])
plt.ylim(-1.0,1.0)
plt.yticks([-1,0,1],[r'$-1$',r'$0$',r'$+1$'])
plt.title(r'$y=\sin(x)$ and $y=\cos(x)$ over $[-\pi, \pi]$')
plt.legend((r'$\cos(x)$', '$\sin(x)$'), loc='upper left')
plt.figtext(0.77,0.73,'This is $\sin(x)$')
plt.figtext(0.25,0.6,'This is $\cos(x)$')
plt.savefig("my-figure.png", dpi=72) #Saves the figure in a local directory
```
